```python
import numpy as np
import pandas as pd
```

```python
# creat a list includes all states abbreviation
statelist = ['AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 
             'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME', 'MI', 'MN', 
             'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM', 'NV', 'NY', 'OH', 
             'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']
```

```python
# read the cost files for wind, solar, and conventional
wind = pd.read_csv('wind_cost_2000to2016.csv')
solar = pd.read_csv('solar_cost_1997to2016.csv')
conventional = pd.read_csv('conventional_cost_2008to2016.csv')
```

```python
# choose the specific year 2015 from dataframe, then drop the unnamed columns and reset index
wind_15 = wind[wind['Year'].isin([2015])].drop(['Unnamed: 0', 'Year'], axis = 1).reset_index(drop = True)
solar_15 = solar[solar['Year'].isin([2015])].drop(['Unnamed: 0', 'Year'], axis = 1).reset_index(drop = True).rename(columns = {'Cost($/MWh)':'solar', 'States':'State'}) 
conventional_15 = conventional[conventional['Year'].isin([2015])].drop(['Unnamed: 0', 'Year'], axis = 1).reset_index(drop = True).rename(columns = {'States':'State'})
```

```python
# creat a new cost dataframe includes average cost for all types of resources by states in 2015
df_cost = pd.DataFrame()
df_cost['State'] = statelist
df_cost.insert(0, 'Year', 2015)
df_cost = pd.merge(df_cost, wind_15, on = 'State', how = 'left')
df_cost = pd.merge(df_cost, solar_15, on = 'State', how = 'left')
df_cost = pd.merge(df_cost, conventional_15, on = 'State', how = 'left').fillna(0)
df_cost.insert(4, 'Hydro', 8.5)
df_cost.rename(columns = {'WindCost':'Wind', 'solar':'Solar'})
```

```python
# export cost dataframe as csv file
df_cost.to_csv('df_cost.csv', index = False)
```
