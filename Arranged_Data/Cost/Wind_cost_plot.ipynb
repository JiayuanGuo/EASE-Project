```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import itertools as it
import math
%matplotlib inline
```

```python
# read the wind cost file and drop the unnamed column
wind = pd.read_csv("wind_cost_2000to2016.csv")
wind = wind.drop(['Unnamed: 0'],axis = 1)
wind.head()
```

```python
# creat a list includes all states abbreviation
state_list = ['AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 
         'HI', 'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 
         'ME', 'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 
         'NJ', 'NM', 'NV', 'NY', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 
         'SD', 'TN', 'TX', 'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']
```

```python
# make the scatter plot of wind cost vs year for each state
for i in state_list:
    plt.figure(figsize = (6,5))
    plt.subplot(111)
    plt.scatter(list(wind[ wind.State == i ][ pd.isnull(wind.WindCost) == False].Year),
                list(wind[ wind.State == i ][ pd.isnull(wind.WindCost) == False].WindCost))
    # set up the x,y labels, and plot title as state name
    plt.ylabel('wind cost')
    plt.xlabel('year')
    plt.title(i)
    plt.grid()
```
