```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load the esource data and drop unnamed column
eperc = pd.read_csv('esource_percent.csv')
eperc = eperc.drop(['Unnamed: 0'], axis = 1)
```

```python
%%javascript
IPython.OutputArea.prototype._should_scroll = function(lines) {
    return false;
}
```

```python
# creat a list contains all states abbreviation
state_list = [
    'AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DC', 'DE', 'FL', 'GA', 'HI', 'IA', 'ID', 'IL', 'IN', 'KS', 
    'KY', 'LA', 'MA', 'MD', 'ME', 'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM', 'NV', 
    'NY', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']
# make the plot of percentage vs year for three different types of clean resources by states
for i in state_list:
    plt.figure(figsize = (10,5))
    plt.subplot(121)
    plt.plot(eperc[eperc.State == i].Year, eperc[eperc.State == i].Hydro, label = 'Hydro')
    plt.plot(eperc[eperc.State == i].Year, eperc[eperc.State == i].Solar, label = 'Solar')
    plt.plot(eperc[eperc.State == i].Year, eperc[eperc.State == i].Wind, label = 'Wind')
    # set up x, y labels and show title as state name abbreviation
    plt.xlabel('Year')
    plt.ylabel('%')
    plt.title(i)
    plt.grid()
    plt.legend()
    plt.subplot(122)
    plt.plot(eperc[eperc.State == i].Year, eperc[eperc.State == i].Solar, label = 'Solar', color = 'C1')
    plt.plot(eperc[eperc.State == i].Year, eperc[eperc.State == i].Wind, label = 'Wind', color = 'C2')
    plt.xlabel('Year')
    plt.ylabel('%')
    plt.title(i)
    plt.grid()
    plt.legend()
```
