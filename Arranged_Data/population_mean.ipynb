```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('arranged_esource')
df = df.drop(['Unnamed: 0', 'Nuclear', 'Other Biomass', 'Other Gases', 'Pumped Storage', 'Total', 'Wood and Wood Derived Fuels', 'Geothermal', 'Other'],axis = 1)
df = df.rename(columns = {'Petroleum':'Petro','Natural Gas':'NG','Hydroelectric Conventional':'Hydro','Solar Thermal and Photovoltaic':'Solar'})
df = df[['Year', 'State', 'Coal', 'NG', 'Petro', 'Hydro', 'Solar', 'Wind']]
df = df[df['Year'].isin([2015])].reset_index(drop = True)
df
```

```python
df_Coal = np.sum(df.Coal)
df_Coal = df_Coal/48
df_NG = np.sum(df.NG)
df_NG = df_NG/50
df_Petro = np.sum(df.Petro)
df_Petro = df_Petro/51
df_Solar = np.sum(df.Solar)
df_Solar = df_Solar/(51-17)
df_Wind = np.sum(df.Wind)
df_Wind = df_Wind/(51-12)
df_Hydro = np.sum(df.Hydro)
df_Hydro = df_Hydro/(48)
```

```python
population_mean = pd.DataFrame()
population_mean.insert(0, 'Year', 2015)
population_mean['Coal'] = df_Coal
population_mean['NG'] = df_NG
population_mean['Petro'] = df_Petro
population_mean['NG'] = df_NG
```
