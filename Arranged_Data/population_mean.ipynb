```python
import numpy as np
import pandas as pd
```

```python
# load average plant capacity file
df = pd.read_csv('average_plant_capacity.csv')
df
```

```python
# calculate the population mean for each type of resources in 2015
df_Coal = np.sum(df.Coal)/47
df_NG = np.sum(df.NG)/48
df_Petro = np.sum(df.Petro)/49
df_Solar = np.sum(df.Solar)/48
df_Wind = np.sum(df.Wind)/44
df_Hydro = np.sum(df.Hydro)/49
mean_list = [df_Coal, df_NG, df_Petro, df_Hydro, df_Solar, df_Wind]
mean_list
```

```python
# creat a dataframe to show population mean for each type of resource in 2015, and convert the type of mean value to integer
population_mean = pd.DataFrame()
row = pd.Series(mean_list,['Coal', 'NG', 'Petro', 'Hydro', 'Solar', 'Wind'])
population_mean = population_mean.append([row]).astype(int)
population_mean
```

```python
population_mean.to_csv('population_mean.csv', index = False)
```
