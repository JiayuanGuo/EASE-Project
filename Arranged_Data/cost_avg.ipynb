```python
data = pd.read_csv('Fossil_Gen_Cost_df.csv')
data = data[data.State != 'District Of Columbia : electric utility']
data = data.sort(['YYMM','State'])
data = data.reset_index(drop=True)
data = data.replace('--',0)
data['Coal','NG','Petroleum'] = data['Coal','NG','Petroleum'].replace('NM',0)
data.Coal = data.Coal.astype(float)
data.NG = data.NG.astype(float)
data.Petroleum = data.Petroleum.astype(float)
data.YYMM = data.YYMM.apply(str)
data.replace(0,np.nan)
```

```python
year = np.arange(2008, 2017)
state = ['AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME', 'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM', 'NV', 'NY', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']
df_avg = pd.DataFrame(columns = ['Year', 'States', 'Coal', 'NG', 'Petro'])
for i in range(len(year)):
    for j in range(len(state)):
        df_avg.loc[i*len(state)+ j] = [str(year[i]), state[j]
                                    , np.nanmean(data[data['YYMM'].str.contains(str(year[i]))][data.States == state[j]].Coal)
                                    , np.nanmean(data[data['YYMM'].str.contains(str(year[i]))][data.States == state[j]].NG)
                                    , np.nanmean(data[data['YYMM'].str.contains(str(year[i]))][data.States == state[j]].Petroleum)]
```
