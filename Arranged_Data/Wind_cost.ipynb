```python
west_list = ['WA','OR','CA','AZ','NV','UT','ID']
interior_list = ['MT','WY','CO','NM','ND','SD','NE','KS','OK','TX','MN','IA','MO']
greatlakes_list = ['WI','IL','IN','OH','MI']
northeast_list = ['PA','NY','ME','VT','NH','MA','RI','CT','NJ']
southeast_list = ['FL','LA','MS','AL','GA','SC','AR','TN','NC','KY','WV','VA','MD','DE']
```

```python
#Extract the cost of different area as dataframe
west_cost = data[data['Area'].isin(['West'])]
interior_cost = data[data['Area'].isin(['Interior'])]
greatlakes_cost = data[data['Area'].isin(['GreatLakes'])]
northeast_cost = data[data['Area'].isin(['Northeast'])]
southeast_cost = data[data['Area'].isin(['Southeast'])]
```

```python
#Represent the cost data in each state by data of corresponding area
wind_cost_df = pd.DataFrame() 
for items in west_list:
    items = west_cost.replace('West',items)
    wind_cost_df = wind_cost_df.append(items)
for items in interior_list:
    items = interior_cost.replace('Interior',items)
    wind_cost_df = wind_cost_df.append(items)
for items in greatlakes_list:
    items = greatlakes_cost.replace('GreatLakes',items)
    wind_cost_df = wind_cost_df.append(items)
for items in northeast_list:
    items = northeast_cost.replace('Northeast',items)
    wind_cost_df = wind_cost_df.append(items)
for items in southeast_list:
    items = southeast_cost.replace('Southeast',items)
    wind_cost_df = wind_cost_df.append(items)
wind_cost_df.head()
```

```python
#Rename the columns, sort, reset index, export the csv file
wind_cost_df.columns = ['Year','State','WindCost']
wind_cost_df = wind_cost_df.sort(['Year','State'])
wind_cost_df.reset_index(inplace=True)
wind_cost = wind_cost_df.iloc[:,1:4]
wind_cost.to_csv('wind_cost_2000to2016.csv')
wind_cost
```
